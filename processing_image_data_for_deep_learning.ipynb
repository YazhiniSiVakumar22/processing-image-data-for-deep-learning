Resizing the image using  pillow library


```python
from PIL import Image
```

```python
img= Image.open('/content/puppy.jpg')
img_resized=img.resize((200,200))
```

```python
img_resized.save('puppy_image_resized.jpg')
```

```python
# displaying the  image from numpy array

img_res =mpimg.imread('/content/puppy_image_resized.jpg')
img_plot=plt.imshow(img_res)
plt.show()
```

```python
print(img_res.shape)
```

Convert RGB images to Grayscale image using OPENCV

```python
# importing opencv library
import cv2
```

```python
img = cv2.imread('/content/puppy.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```

cv2.imshow() will display the image. but this  will not be allowed  in google colab


from google.colab.patches import cv2_imshow

```python
from google.colab.patches import cv2_imshow
```

```python
# displaying the image
cv2_imshow(grayscale_image)
```

```python
# save the grayscale image

cv2.imwrite('puppy_grayscale_image.jpg' ,grayscale_image)
```

```python
# grayscale image resized

#img2= cv2_imshow.open('/content/puppy_grayscale_image.jpg')
img_grayscale_resize=cv2.resize(grayscale_image,(200,200))



```

```python
#showing gray scale resize image
cv2_imshow(img_grayscale_resize)
```

```python
# saving grayscale resize image

cv2.imwrite('puppy_grayscale_resize.jpg' ,img_grayscale_resize)
```
